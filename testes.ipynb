```python
import pandas as pd

df = pd.read_csv('pizzas.csv')
df
```

```python
df.plot(kind='scatter', x='diametro', y='preco', title='Pizza Prices')
```

```python
from sklearn import linear_model

modelo = linear_model.LinearRegression()
x = df[['diametro']]
y = df['preco']

# treinar o modelo
modelo.fit(x, y)
```

```python
print(modelo.predict([[12]])[0])
```
